```python
import numpy as np
import pandas as pd
import os
```

Agrupa todos os dados em um só e exporta um compilado

```python
files = [file for _, _, file in os.walk("./dados/")][0]
dados = [p for p in files if p.endswith("csv")]
dados
```

```python
focos = pd.DataFrame()
for path in dados:
    focos = pd.concat([pd.read_csv("./dados/"+path), focos])
```

```python
focos
```

```python
focos = focos.sort_values(by="datahora")
```

```python
focos = focos.dropna(subset=["frp"])
```

```python
focos.to_csv("Focos.csv", index=None)
```
